```python
class animal:
    '''De Omar Vidaña'''
    name: str

    def __init__(self, name):
        self.name = name

    def eat(self):
        print(self.name + " está comiendo.")

    def hug(self):
        print(self.name + " no es abrazable")


class hugable(animal):
    def __init__(self, name):
        tipo = name + " Abrazable"
        super().__init__(tipo)

    def hug(self):
        print("Abrazamos a " + self.name)


class duck(hugable):
    def __init__(self, name):
        tipo = "Pato " + name
        super().__init__(tipo)

    def sound0(self):
        print("Cuac Cuac")


class hedgehog(animal):
    pass
```

```python
Hewey = duck("Hewey")
Hewey.eat()
Hewey.hug()
Hewey.sound0()
print(Hewey)
Butch = hedgehog("Butch")
Butch.hug()
```

```python
lista = (("Hewey", "duck"),
         ("Dewey", "duck"),
         ("Lewie", "duck"),
         ("Butch", "hedgehog"),
         ("Chuck", "hedgehog"),
         ("Sparky", "dog"),
         ("blue", "dog"))

animals = []
for name, tipo in lista:
    if tipo == "duck":
        animals.append( duck(name) )
    elif tipo == "hedgehog":
        animals.append( hedgehog(name) )
    elif tipo == "dog":
        animals.append( hugable(name) )

for animal in animals:
    animal.hug()

print ("\n")
        
```

## Ejercicio

### 1. Situación:  

Se tiene una empresa con programadores, secretarias y gerentes.
Los programadores y las secretarias reciben un sueldo base más un bono por antigüedad de \$1000 por año trabajado.
Los gerentes sólo reciben su salario.
Los gerentes pueden cambiar el salario de los programadores en cualquier momento, pero sólo para incrementarlo.
Todos tienen una función trabajar, pero el gerente gerencia; el programador programa y la secretaria apoya. Para subirlos al sistema, se conoce la antigüedad, el nombre y el puesto de la persona.
Las clases que modelen el sistema deben incluir un método que calcule el salario adecuadamente y lo regrese para imprimirlo.


Salarios base:

- Gerentes: __\$50,000__  

- Secretarias: __\$15,000__   

- Programadores: __\$25,000__  

    

### 1. Objetivos    

    1. Modela las clases necesarias para representar esta situación.

    2. Programar las clases utilizando herencia

    3. Crear una prueba en la que se generen uno de cada tipo de empleado y se utilice por lo menos una vez cada función.

```python
class worker():
  '''De Omar Vidaña'''
  name: ""
  pos: ""
  antiquity: 0
  base_salary: 0.0

  def __init__(self, name, position, antiquity, base_salary):
    self.name = name
    self.pos = position
    self.antiquity = antiquity
    self.base_salary = base_salary
    
  def work(self):
    print(f"{self.name} está trabajando")
    
  def get_salary(self):
    print(f"Salario: {self.base_salary}")
      
  def __str__(self):
      return f"Puesto: {self.pos} - Nombre: {self.name} - Antigüedad: {self.antiquity} - Salario base: {self.base_salary}"
				
class manager(worker):
  def __init__(self, name, position, antiquity):
      super().__init__(name, position, antiquity, 50000)
    
  def change(self, employee):
    if isinstance(employee, programmer):
      print(f"Ingrese el nuevo salario base")
      base = float(input())
      if employee.base_salary >= base:
        print(f"El nuevo salario debe ser mayor al anterior: ({employee.base_salary})")
      else:
        employee.base_salary = base
        print(f"El nuevo salario ha sido actualizado")
    else:
      print(f"Solamente se puede modifcar el salario de un programador.")
        
  def work(self):
    print(f"{self.name} está Gerenciando")

class programmer(worker):
  def __init__(self, name, position, antiquity):
    super().__init__(name, position, antiquity, 25000)
  
  def work(self):
    print(f"{self.name} está Programando")
  
  def get_salary(self):
    print(f"Salario: {self.base_salary + (self.antiquity * 1000)}")
  
class secretary(worker):
  def __init__(self, name, position, antiquity):
      super().__init__(name, position, antiquity, 15000)
    
  def work(self):
    print(f"{self.name} está Apoyando")
  
  def get_salary(self):
    print(f"Salario: {self.base_salary + (self.antiquity * 1000)}")

Mario = secretary("Mario Vidaña", "Secretario", 4)
print(Mario)
Mario.get_salary()
Juan = manager("Juan", "Manager", 10)
print(Juan)
Juan.get_salary()
Julieta = programmer("Julieta", "Programador", 3)
print(Julieta)
Julieta.get_salary()

Mario.work()
Juan.work()
Julieta.work()

Juan.change(Mario)
Juan.change(Julieta)

print(Mario)
Mario.get_salary()
print(Juan)
Juan.get_salary()
print(Julieta)
Julieta.get_salary()
```
